## 3. ***Data Wrangling***

Data wrangling (also called data cleaning or data preprocessing) is the process of transforming raw data into a clean, structured, and usable format,

 Goal of Data Wrangling : Make the dataset clean, reliable, and analysis-ready, so insights and patterns can be found easily.

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

# 1️.Remove Duplicate Rows
combine_dataset.drop_duplicates(inplace=True)

# 2️.Handle Missing Values
# We can either drop or fill missing values depending on the context

# Example: Fill missing values with a default or placeholder
combine_dataset['age_certification'].fillna('UNKNOWN', inplace=True)
combine_dataset['seasons'].fillna(0, inplace=True)
combine_dataset['description'].fillna('No Description Available', inplace=True)

# 3️.Convert data types if needed
combine_dataset['seasons'] = combine_dataset['seasons'].astype(int)
combine_dataset['release_year'] = combine_dataset['release_year'].astype(int)
combine_dataset['runtime'] = combine_dataset['runtime'].astype(int)

# 4️.Strip extra spaces and lowercase for consistency
combine_dataset['type'] = combine_dataset['type'].str.strip().str.lower()
combine_dataset['role'] = combine_dataset['role'].str.strip().str.upper()
````

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

Data Visualization is the graphical representation of information and data using charts, graphs, and plots. It helps us see patterns, trends, and outliers in the data.
There are 3 types of Data Visualization such as :

1️. Univariate Analysis (Uni = One) - Looking at one variable at a time.

Purpose: Understand the basic distribution, patterns, or characteristics of a single column.

Example : What’s the average IMDb Score?
          How many titles fall under each Genre?
          Common Visuals: Histogram, Pie chart, Bar chart, Count plot


2️. Bivariate Analysis (Bi = Two) - Studying the relationship between two variables.

Purpose: Find how one variable affects or is related to another.

Example : Does runtime vary with type (movie vs show)?
          Does IMDb Score change with Age Certification?
          Common Visuals: Scatter plot, Box plot, Bar plot, Heatmap, Correlation matrix


3️. Multivariate Analysis (Multi = More than Two) - Examining more than two variables together.

Purpose: Detect deeper patterns, interactions, or trends involving 3 or more factors.

Example : How do IMDb Score, Genre, and Runtime together affect a title’s popularity?
          Do Genre, Age Certification, and Type influence TMDB Score?
          Common Visuals: Pair plots, Multicolored scatter plots, Heatmaps, 3D charts
